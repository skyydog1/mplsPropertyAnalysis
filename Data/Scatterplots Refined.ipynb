```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas
```

```python
data = pd.read_csv('Assessors_Parcel_Data_2022.csv', low_memory=False)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = data[['X','Y','WARD','LANDVALUE','BLDGVALUE','TOTALVALUE','LANDUSE','PARCEL_AREA_SQFT','BUILDINGUSE','YEARBUILT']]
```

```python
df = df[df['TOTALVALUE'] != 0] #remove 0 values for ln
```

```python
df['color_array'] = np.log(df['TOTALVALUE']) #create color array dataframe column (natural log of total value)
fig, ax = plt.subplots(figsize=(30, 40))
ax.scatter(df['X'], df['Y'],s = 10, c = df['color_array'], cmap = 'viridis') #This plot works by mapping the NAD83 coordinates of each house and then coloring by totalvalue
```
